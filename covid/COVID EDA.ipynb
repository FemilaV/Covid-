# John Hopkins Covid 19 dataset
## There are 3 datasets
- Confirmed cases of all countries
- death cases of all countries
- recovered cases of all countries


```python
#importing pandas and other libraries required for EDA
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("ggplot")
import warnings
warnings.filterwarnings("ignore")
```

```python
#reading all the datasets
```

```python
df = pd.read_csv('time_series_covid19_confirmed_global.csv',index_col='Country/Region')
df_d = pd.read_csv('time_series_covid19_deaths_global.csv',index_col='Country/Region')
df_r = pd.read_csv('time_series_covid19_recovered_global.csv',index_col='Country/Region')
```

```python
df.head(20)
```

```python
columns = df.columns
columns
```

```python
columns_d = df_d.columns
columns_d
```

```python
columns_r = df_r.columns
columns_r
```

```python
#droping Latitude and longitude columns as finding it irrelevant for the further steps
df.drop(columns = ['Lat','Long'], axis =1, inplace = True)
df_d.drop(columns = ['Lat','Long'], axis =1, inplace = True)
df_r.drop(columns = ['Lat','Long'], axis =1, inplace = True)
```

## Grouping datasets by country name

```python
# Grouping the datasets as country names as dataset is having details of some countries different provinces covid 19 details,
#like Australia's 8 provinces are given seperately,want to deal with it as a single column

covid = df.groupby("Country/Region").sum()
covid.head(50)
```

```python
covid.shape
```

```python
# taking the mostly affected countries to plot
most_affected=covid.sort_values(by = columns[-1],ascending = False)
most_affected=most_affected.head(15)
```

```python
#finding the least affected countries
least_affected=covid.sort_values(by = columns[-1],ascending = True)
least_affected=least_affected.head(15)
```

```python
covid_d = df_d.groupby("Country/Region").sum()
covid_d.head()
```

```python
most_d=covid_d.sort_values(by = columns[-1],ascending = False)
most_d=most_d.head(15)
```

```python
least_d=covid_d.sort_values(by = columns[-1],ascending = True)
least_d=least_d.head(15)
```

```python
covid_r = df_r.groupby("Country/Region").sum()
covid_r.head()
```

```python
most_r=covid_r.sort_values(by = columns[-1],ascending = False)
most_r=most_r.head(15)
```

```python
least_r=covid_r.sort_values(by = columns[-1],ascending = True)
least_r=least_r.head(15)
```

# Plot Highly effected ,high no of death cases and highly recovered cas countries

```python
plt.figure(figsize=(9,10))
ax = sns.barplot(x=most_affected[most_affected.columns[-1]], y=most_affected.index.values,palette = "Accent_r")
ax.set_title("Highly affected countries")
```

```python
plt.figure(figsize=(9,10))
ax = sns.barplot(x=most_d[most_d.columns[-1]], y=most_d.index.values,palette ="Blues_d")
ax.set_title("Countries with higher death cases")
```

```python
plt.figure(figsize=(9,10))
ax = sns.barplot(x=most_r[most_r.columns[-1]], y=most_r.index.values)
ax.set_title("Countries with higher recovery")
```

# Plot least affected ,low no: of deaths and less recovered countries

```python
plt.figure(figsize=(9,10))
ax = sns.barplot(x=least_affected[least_affected.columns[-1]], y=least_affected.index.values,palette = "Blues_d")
ax.set_title("Least affected countries")
```

```python
plt.figure(figsize=(9,10))
ax = sns.barplot(x=least_d[least_d.columns[-1]], y=least_d.index.values,palette = "Blues_d")
ax.set_title("Countries with least number of deaths")
```

```python
plt.figure(figsize=(9,10))
ax = sns.barplot(x=least_r[least_r.columns[-1]], y=least_r.index.values,palette = "Accent_r")
ax.set_title("Countries with least number of recovery")
```

# Swapping the rows and columns,to get the dates as index column

```python
#transposing the rows and columns so that we will get the dates as the index and different countries as the columns
covid_trans =covid.transpose()
```

```python
covid_r = covid_r.transpose()
```

```python
covid_r
```

## changing date column to datetime format

```python
from datetime import datetime
```

```python
covid_trans['index'] = pd.to_datetime(covid_trans['index'],infer_datetime_format=True)
```

```python
covid_trans.tail()
```

```python
covid_trans.head()
```

```python
covid_d['index'] = pd.to_datetime(covid_d['index'],infer_datetime_format=True)
```

```python
covid_r['index'] = pd.to_datetime(covid_r['index'],infer_datetime_format=True)
```

# Plot total covid cases of whole world in each month

```python
count = []
for i in range(0,243):
    count.append(sum(pd.to_numeric(covid_trans.iloc[i,1:].values)))

df = pd.DataFrame()
df["dates"] = covid_trans['index'][1:]
df["Cases"] = count
df=df.set_index("dates")


df.Cases.plot(title="Daily Covid19 Cases in World",marker="+",figsize=(10,5),label="daily cases")
df.Cases.rolling(window=5).mean().plot(figsize=(10,5),label="MA5")
plt.ylabel("Cases")
plt.legend()
plt.show()
```

# Plot total death cases of whole world in each month

```python
count = []
for i in range(0,243):
    count.append(sum(pd.to_numeric(covid_d.iloc[i,1:].values)))

df["Deaths"] = count
```

```python
df.Deaths.plot(title="Daily Covid19 Deaths in World",marker=".",figsize=(10,5),label="daily deaths")
df.Deaths.rolling(window=5).mean().plot(figsize=(10,5),label="MA5")
plt.ylabel("Deaths")
plt.legend()
plt.show()
```

# Plot total recovery cases of whole world in each month

```python
count = []
for i in range(0,243):
    count.append(sum(pd.to_numeric(covid_r.iloc[i,1:].values)))

df["Recovery"] = count
```

```python
df.Deaths.plot(title="Daily Covid19 Recovery in World",marker=".",figsize=(10,5),label="daily recovery")
df.Deaths.rolling(window=5).mean().plot(figsize=(10,5),label="MA5")
plt.ylabel("Recovery")
plt.legend()
plt.show()
```
